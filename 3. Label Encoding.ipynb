```python
import pandas as pd
import datetime
```

```python
df_base=datetime.datetime.today()
df_base
```

```python
df_date_list=[df_base - datetime.timedelta(days=x) for x in range(0,20)]
df_date_list
```

```python
df=pd.DataFrame(df_date_list)
df
```

```python
df.columns=['day']
df
```

```python
df['day_of_week']=df['day'].dt.day_name()
df['day_of_week']
```

```python
df.head()
```

```python
weekday_map={
    'Monday':'1',
    'Tuesday':'2',
    'Wednesday':'3',
    'Thursday':'4',
    'Friday':'5',
    'Saturday':'6',
    'Sunday':'7'
    
}
```

```python
df['day_ordinal']=df.day_of_week.map(weekday_map)
```
