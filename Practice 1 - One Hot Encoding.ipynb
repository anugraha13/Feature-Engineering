# Label Encoding for a date

```python
from datetime import datetime, timedelta
```

```python
current=datetime.now()
```

```python
current
```

```python
datelist=[current-timedelta(days=x) for x in range(0,20)]
```

```python
datelist
```

```python
df=pd.DataFrame(datelist)
```

```python
df.drop('columns',axis=1,inplace=True)
```

```python
df.columns=['Day']
```

```python
df
```

```python
df['weekday']=df['Day'].dt.day_name()
```

```python
df.head()
```

```python
day_list={
    'Monday':'1',
    'Tuesday':'2',
    'Wednesday':'3',
    'Thursday':'4',
    'Friday':'5',
    'Saturday':'6',
    'Sunday':'7'
}
```

```python
df['order']=df.weekday.map(day_list)
```

```python
df.head()
```

# target guided


```python
import pandas as pd
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
df.Cabin.fillna('Missing',inplace=True)
```

```python
df.Cabin=df.Cabin.apply(lambda x : str(x[0]))
```

```python
df.Cabin
```

```python
df.head()
```

```python
df.groupby('Cabin')['Survived'].mean().sort_values(ascending=False).index
```

```python
df.Cabin.value_counts()
```

```python
ordinal_labels=df.groupby('Cabin')['Survived'].mean().sort_values(ascending=False).index
```

```python
ordinal_labels
```

```python
mapp={k:i for i,k in enumerate(ordinal_labels,0)}
mapp
```

```python
df['order']=df['Cabin'].map(mapp)
```

```python
df.head()
```
