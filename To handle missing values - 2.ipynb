# Random Sample imputation

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

### Aim : Taking random observation from the dataset and replace the NAN values based on the observation did with random samples

### When to be used : Data are missing completely at Random

```python
df.isnull().sum()
```

```python
df.isnull().mean() # percentage of null values
```

```python
df['Age'].isnull().sum()
```

```python
pd.set_option('display.max_rows',None)
df[df['Age'].isnull()]
```

```python
test=df['Age'].dropna().sample(df['Age'].isnull().sum())
test
```

```python
len(test)
```

```python
df[df['Age'].isnull()].index
```

```python
def replace_nan(df,variable,median):
    df[variable+'_Median']=df[variable].fillna(median)
    df[variable+'_random']=df[variable]
    random_sample=df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    random_sample.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random']=random_sample
```

```python
median=df.Age.median()
```

```python
replace_nan(df,'Age',median)
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
df['Age'].plot(kind='kde',ax=ax)
df.Age_random.plot(kind='kde',ax=ax,color='green')
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

### Advantages : Easy to implement, less distortion in variance, 
### Disadvantages : Every situaiton, randomness wont work

# End of Distribution Imputation

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.Age.hist(bins=50)
```

```python
extreme=df.Age.mean()+3*df.Age.std() # beyond the 3rd normal distribution
```

```python
import seaborn as sns
sns.boxplot('Age',data=df) # finding outliers
```

```python
def replace_nan(df,variable,median,extreme):
    df[variable+'_end_distribution']=df[variable].fillna(extreme)
    df[variable].fillna(median,inplace=True)
# replaced null age column with median value
```

```python
replace_nan(df,'Age',df.Age.median(),extreme)
```

```python
df.head()
```

```python
import seaborn as sns
sns.boxplot('Age_end_distribution',data=df)
```

```python
import seaborn as sns
sns.boxplot('Age',data=df)
```

```python
df.Age.hist(bins=50) # large distortion in variance since this was replaced by median -- huge value
```

```python
df.Age_end_distribution.hist(bins=50)
```

```python
# Advantages - Easy to implement
# Disadvantages - if the number of NA is large, it masks the true outliers, 
# if the number of NA is small, then it will be considered as outliers and need to be pre-processed further
# distorts the original distribution
```
