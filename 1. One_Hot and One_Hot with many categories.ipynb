```python
def one_hot_top(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
```

```python
one_hot_top(df,'X2',top10)
df.head()
```

```python
top20=[x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]
```

```python
one_hot_top(df,'X1',top20)
df.head()
```
